**Tratamiento de la columna int_date**

En primer lugar dividimos la columna int_date en tres columnas:

```python
df_work[["int_day", "int_month", "int_year"]] = df_work['int_date'].str.split("-", expand= True)
df_work.head()
```

Sustituimos el mes en letra por el número de mes:

```python
#Paso 1: Creamos un diccionario de tuplas con el mes y el número
meses = {
    'enero': '01',
    'febrero': '02',
    'marzo': '03',
    'abril': '04',
    'mayo': '05',
    'junio': '06',
    'julio': '07',
    'agosto': '08',
    'septiembre': '09',
    'octubre': '10',
    'noviembre': '11',
    'diciembre': '12'
}
#Paso 2 --> Utilizamos la función map para cambiar el nombre del mes por su número: 
df_work['int_month'] = df_work['int_month'].map(meses)
#Paso 3 --> Observamos el resultado obtenido: 
df_work['int_month']
    
```

Ahora convertimos la columna int_day en una columna con dos dígitos.

```python
#1. Creamos una lista vacía en la que volcaremos los resultados
days = []
#2. Creamos un bucle que recorra toda la columna:
for x in df_work['int_day']:
    if pd.isna(x):
         days.append(np.nan)  #Si encuentra un nulo lo devuelve como tal
    else:
        if int(x) < 10:
            days.append ("0" + str(x)) #Si el numero es menor que 10 le añade un 0
        else:
            days.append(str(x)) #Si no lo devuelve como tal 
df_work['int_day'] = days #Volcamos el resultado a la columna original

df_work['int_day'] 
            
```

El último ajuste que hacemos es concatenar las fechas en una nueva columna.

```python
#Vemos la cantidad de nulos en la columna int_date
nulos(df_work['int_date'])
```

```python
#Creamos una lista vacía en la que volcaremos los nuevos valores de la columna int_date
Nint = []
#Creamos una función que si encuentra nulos devuelva el nulo y si no la fecha concatenada: 
for i, x in enumerate(df_work['int_date']): #Accedemos a cada uno de los valores de la columna int_date
    #Si es nulo lo adjuntamos como nulo 
    if pd.isna(x):
        Nint.append(np.nan)
    else:
        #Accedemos a cada valor de las columnas day, month, year y los convertimos a string
        day = str(df_work.loc[i, 'int_day'])
        month = str(df_work.loc[i, 'int_month'])
        year = str(df_work.loc[i, 'int_year'])
        #Concatenamos los valores
        Nint.append(day + "-" + month + "-" + year)

#Convertimos la columna a formato datetime 
fechas = [] #Creamos de nuevo una lista vacía donde volcaremos las fechas 
for x in Nint:#Creamos un bucle que devuelva nulos y fechas
    if pd.isna(x):
        fechas.append(np.nan)
    else:
        fechas.append(pd.to_datetime(x, format = '%d-%m-%Y'))

#Volcamos la lista en una nueva columna del DF
df_work['new_int_date'] = fechas
#Verificamos el resultado y la cantidad de nulos para ver que esta no ha variado. 
print(df_work['new_int_date'])
nulos(df_work['new_int_date'])
```

Como vemos hay valores que no son realistas en los registros ya que una llamada de 0 segundos no tiene sentido a la vez que una llamada de 82 minutos tampoco lo tiene.
```python
mayores_20 = (df_work['duration'] > 20).sum()
menores_1 = (df_work['duration'] < 1).sum()
valores_a_cambiar = mayores_20 + menores_1
print (f'Hay {mayores_20} registros superiores a 20 mins y {menores_1} registros inferiores a 1 min')
print (f'Vamos a cambiar {valores_a_cambiar} valores')
```

```python
#Creamos una función lambda para hacer el cambio
df_work['duration'] = df_work['duration'].apply(lambda x: np.nan if x > 20 or x < 1 else x)
print("Se ha aplicado la corrección")
```
