Dataset customer details

```python
df_cust_sucio = pd.read_csv("../data/customer-details.csv")
df_cust_sucio.head()
```

DESCRIPCIÓN DE COLUMNAS

- Income: Representa el ingreso anual del cliente en términos monetarios.
- Kidhome: Indica el número de niños en el hogar del cliente.
- Teenhome: Indica el número de adolescentes en el hogar del cliente.
- Dt_Customer: Representa la fecha en que el cliente se convirtió en cliente de la empresa.
- NumWebVisitsMonth: Indica la cantidad de visitas mensuales del cliente al sitio web de la empresa.
- ID: Identificador único del cliente.

A continuación, vamos a combinar ambos datasets en uno solo utilizando un inner join en la columna ID para ello vamos a normalizar el nombre las columnas id de ambos sets:

CAMBIO 1:
Normalizamos el nombre de la columna ID a id del dataset de customer-details

```python
df_cust_sucio = df_cust_sucio.rename(columns={'ID':'id'})
df_cust_sucio.columns
```

CAMBIO 2:
Normalizamos el nombre de la columna id_ a id del dataset bank-additional

```python
df_bank_sucio = df_bank_sucio.rename(columns={'id_':'id'})
df_bank_sucio.columns
```

Una vez normalizadas ambas columnas vamos a crear un único dataset utilizando un innerjoin en la columna id de ambos sets

```python
df_definitivo_sucio = pd.merge(df_bank_sucio, df_cust_sucio, on= 'id', how= 'inner')
df_definitivo_sucio.head()
```

Exportamos el nuevo dataset como un archivo csv:

```python
df_definitivo_sucio.to_csv("data_definitivo", index= False)
```
